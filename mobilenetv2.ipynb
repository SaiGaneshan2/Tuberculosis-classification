```python
import tensorflow as tf
import cv2
import numpy as np
import pathlib
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tqdm import tqdm 

train_path = pathlib.Path("/home/sg/Music/TB DATASET/DATASET/train")
aug_train_path = pathlib.Path("/home/sg/Music/TB DATASET/DATASET/aug_train_dic")
aug_train_path.mkdir(parents=True, exist_ok=True)

datagen = ImageDataGenerator(
    rotation_range=10,         # Moderate rotation
    width_shift_range=0.1,     # Less horizontal shift
    height_shift_range=0.1,    # Less vertical shift
    shear_range=0.1,          # Moderate shear
    zoom_range=0.1,           # Less zoom effect
    horizontal_flip=True,     # Flip images
    fill_mode="nearest"
)

# Desired number of images per class
TARGET_COUNT = 13908

for category in os.listdir(train_path):
    category_path = train_path / category
    aug_category_path = aug_train_path / category
    aug_category_path.mkdir(parents=True, exist_ok=True)

    original_images = os.listdir(category_path)
    original_count = len(original_images)

    print(f"Processing {category}...")

    # Calculate augmentations needed per image
    augmentations_needed = TARGET_COUNT - original_count
    aug_per_image = augmentations_needed // original_count
    extra_augmentations = augmentations_needed % original_count

    for idx, img_name in enumerate(tqdm(original_images)):
        img_path = category_path / img_name
        img = cv2.imread(str(img_path))
        
        if img is None:
            continue
        
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # RGB conversion
        img = tf.keras.preprocessing.image.img_to_array(img)  # To array
        img = np.expand_dims(img, axis=0)  # Add batch dimension

        aug_iter = datagen.flow(img, batch_size=1)

        # Augment 'aug_per_image' times
        for i in range(aug_per_image):
            aug_img = next(aug_iter)[0].astype(np.uint8)
            aug_img_path = aug_category_path / f"aug_{i}_{img_name}"
            cv2.imwrite(str(aug_img_path), cv2.cvtColor(aug_img, cv2.COLOR_RGB2BGR))

        # Distribute extra augmentations if necessary
        if idx < extra_augmentations:
            aug_img = next(aug_iter)[0].astype(np.uint8)
            aug_img_path = aug_category_path / f"aug_extra_{img_name}"
            cv2.imwrite(str(aug_img_path), cv2.cvtColor(aug_img, cv2.COLOR_RGB2BGR))

print("‚úÖ Augmentation Complete! Both classes now have 13,908 images each.")
```

```python
import tensorflow as tf
import cv2
import numpy as np
import pathlib
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tqdm import tqdm 

# Paths
train_path = pathlib.Path("/home/sg/Music/TB DATASET/DATASET/train")
aug_train_path = pathlib.Path("/home/sg/Music/TB DATASET/DATASET/aug_train_dic")
aug_train_path.mkdir(parents=True, exist_ok=True)

# Data augmentation settings
datagen = ImageDataGenerator(
    rotation_range=10,         
    width_shift_range=0.1,     
    height_shift_range=0.1,    
    shear_range=0.1,          
    zoom_range=0.1,           
    horizontal_flip=True,     
    fill_mode="nearest"
)

# Target number of images per class
TARGET_COUNT = 10431  
current_count = 7811  # Current NORMAL image count
additional_needed = TARGET_COUNT - current_count  # 2620 more images needed

# Process only the 'NORMAL' category
category = "NORMAL"
category_path = train_path / category
aug_category_path = aug_train_path / category
aug_category_path.mkdir(parents=True, exist_ok=True)

original_images = sorted(os.listdir(category_path))  # Sort images for consistency
selected_images = original_images[:500]  # Select first 500 images

print(f"Processing first 500 NORMAL images for augmentation...")

generated_images = 0

for idx, img_name in enumerate(tqdm(selected_images)):
    if generated_images >= additional_needed:  # Stop if we have enough images
        break
    
    img_path = category_path / img_name
    img = cv2.imread(str(img_path))
    
    if img is None:
        continue
    
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  
    img = tf.keras.preprocessing.image.img_to_array(img)  
    img = np.expand_dims(img, axis=0)  

    aug_iter = datagen.flow(img, batch_size=1)

    # Generate 3 augmented images per original image
    for i in range(3):  
        if generated_images >= additional_needed:  # Stop when we reach the target
            break
        aug_img = next(aug_iter)[0].astype(np.uint8)
        aug_img_path = aug_category_path / f"aug_{generated_images}_{img_name}"
        cv2.imwrite(str(aug_img_path), cv2.cvtColor(aug_img, cv2.COLOR_RGB2BGR))
        generated_images += 1  # Increment count of generated images

print(f"‚úÖ Augmentation Complete! NORMAL class now has {TARGET_COUNT} images.")
```

```python
IMAGE_SHAPE = (224, 224)
BATCH_SIZE = 16

train_datagen = ImageDataGenerator(rescale=1.0/255)
val_datagen = ImageDataGenerator(rescale=1.0/255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=IMAGE_SHAPE,
    batch_size=BATCH_SIZE,
    class_mode='sparse'  # Sparse labels for classification
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=IMAGE_SHAPE,
    batch_size=BATCH_SIZE,
    class_mode='sparse'
)

num_classes = len(train_generator.class_indices)  # Get the number of classes
```

```python
import tensorflow as tf

# Load Pretrained ResNet101V2 (Without Top Layer)
base_model = tf.keras.applications.ResNet101V2(input_shape=(224, 224, 3), include_top=False, weights="imagenet")

# Freeze all layers first
base_model.trainable = False

# Unfreeze only the last convolutional block (adjustable)
for layer in base_model.layers[-10:]:  # Unfreezing the last 10 layers
    layer.trainable = True

# Define Model
inputs = tf.keras.Input(shape=(224, 224, 3))
x = base_model(inputs, training=False)  # Pass input through ResNet101V2
x = tf.keras.layers.GlobalAveragePooling2D()(x)  # Pooling Layer
x = tf.keras.layers.Dropout(0.3)(x)  # üî• Dropout to prevent overfitting
x = tf.keras.layers.BatchNormalization()(x)  # Normalization to stabilize training
x = tf.keras.layers.Dense(128, activation="relu")(x)  # Fully connected layer
x = tf.keras.layers.Dropout(0.2)(x)  # üî• Additional Dropout layer
outputs = tf.keras.layers.Dense(1, activation="sigmoid")(x)  # ‚úÖ Single neuron for binary classification

model = tf.keras.Model(inputs, outputs)

# Compile the model
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

# Print summary
model.summary()
```
